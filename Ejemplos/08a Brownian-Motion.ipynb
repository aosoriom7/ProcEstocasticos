```python
# https://github.com/mohammadjafariph/Brownian-Motion-Simulation/pulls
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d as p3
from matplotlib.animation import FuncAnimation as animation
%matplotlib inline
```

```python
Npoints = 8192
T = 1.0
xaxis = np.linspace(0.0,T,Npoints)
#print(xaxis[1])
Nparticle = 3
dT = xaxis[1] - xaxis[0]
#print(dT)
dX = np.sqrt(dT) * np.random.randn(Nparticle,Npoints)
# For brownian move, we have Xt2 - Xt1 ~ Normally distributed with mean 0 and variance t2 - t1, generated by various normally distributed increments
X = np.cumsum(dX,axis = 1)

Nbins = int(np.ceil(1 + 3.3*np.log(Npoints))) 
kwargs = dict(histtype='stepfilled', alpha=0.3,  bins=Nbins)
plt.figure(figsize=(9, 3), dpi=120)
for i in range(Nparticle):
    plt.plot(xaxis , X[i,:])

plt.figure(figsize=(9, 3), dpi=120)
for i in range(Nparticle):
    plt.hist(X[i,:], **kwargs);
plt.show()
```

```python
dX = np.sqrt(dT) * np.random.randn(Nparticle,Npoints)
X = np.cumsum(dX,axis = 1)
dY = np.sqrt(dT) * np.random.randn(Nparticle,Npoints)
Y = np.cumsum(dY,axis = 1)
for i in range(Nparticle):
    plt.plot(X[i,:] , Y[i,:])
plt.show()
```

```python
dX = np.sqrt(dT) * np.random.randn(Nparticle,Npoints)
X = np.cumsum(dX,axis = 1)
dY = np.sqrt(dT) * np.random.randn(Nparticle,Npoints)
Y = np.cumsum(dY,axis = 1)
dZ = np.sqrt(dT) * np.random.randn(Nparticle,Npoints)
Z = np.cumsum(dZ,axis = 1)
sp = plt.axes(projection = "3d")
for i in range(Nparticle):
    sp.plot3D(X[i,:] , Y[i,:] , Z[i,:] )
plt.show()
```
