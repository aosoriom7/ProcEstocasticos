```python
# generador de textos mediante cadenas de Markov
# Data - text
import numpy as np
data = open('SpanishText.txt', encoding='utf8').read()

print(data)
```

```python
# Segmenting words
ind_words = data.split()
print(ind_words)
```

```python
# pairing words
def make_pairs(ind_words):
    for i in range(len(ind_words) - 1):
        yield (ind_words[i], ind_words[i + 1])
pair = make_pairs(ind_words)
```

```python
word_dict = {}
for word_1, word_2 in pair:
    if word_1 in word_dict.keys():
        word_dict[word_1].append(word_2)
    else:
        word_dict[word_1] = [word_2]
```

```python
# chooosing pair of words randomly
first_word = np.random.choice(ind_words)
 
while first_word.islower():
    chain = [first_word]
    n_words = 21 
    first_word = np.random.choice(ind_words)
    
    for i in range(n_words):
        chain.append(np.random.choice(word_dict[chain[-1]]))
```

```python
print(' '.join(chain))
```
