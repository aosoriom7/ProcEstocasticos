```python
from matplotlib import pyplot as plt
from pandas import read_csv
from pandas.plotting import lag_plot, autocorrelation_plot
```

```python
# ploting of data
plt.figure(figsize=(9, 6), dpi=120)
ax =  plt.subplot(111)
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0)
print(series.head())
series.plot(ax=ax,title ='daily minimum temperatures',color="midnightblue")
```

```python
plt.figure(figsize=(9, 6), dpi=120)
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0)
lag_plot(series, lag=1, marker='.')
```

```python
plt.figure(figsize=(9, 6), dpi=120)
ax =  plt.subplot(111)
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0)
autocorrelation_plot(series, ax=ax)
```

```python
import numpy as np
ps = np.abs(np.fft.fft(series))**2
time_step = 1 
freqs = np.fft.fftfreq(series.size, time_step)
idx = np.argsort(freqs)
plt.figure(figsize=(9, 6), dpi=120)
plt.semilogy(freqs[idx], ps[idx])
plt.title('Power spectrum (np.fft.fft)')
```

```python
import math
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0)

plt.figure(figsize=(9, 6), dpi=120)
plt.specgram(series['Temp'], Fs=1);
plt.title('Spectrogram Using matplotlib.pyplot.specgram() method')  
plt.show() 
 
```
